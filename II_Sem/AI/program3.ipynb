```python
import sys
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
def main():
    try:
        # Load dataset
        iris = load_iris()
        X = iris.data
        Y = iris.target
        print("Iris dataset loaded successfully!")
        print(f"Feature names: {iris.feature_names}")
        print(f"Target names: {iris.target_names}\n")
        # Train-test split
        X_train, X_test, Y_train, Y_test = train_test_split(
            X, Y, test_size=0.2, random_state=42, stratify=Y
        )
        try:
            k = int(input("Enter the value of k (positive integer): "))
            if k <= 0:
                raise ValueError("k must be a positive integer.")
        except ValueError as e:
            print(f"Invalid input for k: {e}")
            sys.exit(1)
        # Train KNN model
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, Y_train)
        # Predictions
        Y_pred = knn.predict(X_test)
        # Model Evaluation
        print("\nModel Evaluation:")
        print(f"Accuracy: {accuracy_score(Y_test, Y_pred) * 100:.2f}%")
        print("\nClassification Report:")
        print(classification_report(Y_test, Y_pred, target_names=iris.target_names))
        # Sample prediction
        sample = np.array([[5.1, 3.5, 1.4, 0.2]])
        predicted_class = knn.predict(sample)[0]
        print(f"\nExample prediction for {sample.tolist()}: {iris.target_names[predicted_class]}")
    except Exception as e:
        print(f"An error occurred: {e}")
        sys.exit(1)
if __name__ == "__main__":
    main()
```

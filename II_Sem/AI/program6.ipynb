```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import os
def load_dataset(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Dataset file '{file_path}' not found.")
    # Dataset is tab-separated: label \t message
    df = pd.read_csv(file_path, sep='\t', header=None, names=['label', 'message'])
    return df
def preprocess_data(df):
    # Convert labels to binary: ham=0, spam=1
    df['label'] = df['label'].map({'ham': 0, 'spam': 1})
    return df
def train_naive_bayes(X_train, y_train):
    model = MultinomialNB()
    model.fit(X_train, y_train)
    return model

# -------------------------------
# Main Execution
# -------------------------------
if __name__ == "__main__":
    try:
        # Load and preprocess
        dataset_path = "SMSSpamCollection"  # Change path if needed
        df = load_dataset(dataset_path)
        
       df = preprocess_data(df)

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            df['message'], df['label'], test_size=0.2, random_state=42
        )

        # Convert text to numerical features
        vectorizer = CountVectorizer(stop_words='english')
        X_train_vec = vectorizer.fit_transform(X_train)
        X_test_vec = vectorizer.transform(X_test)

        # Train model
        model = train_naive_bayes(X_train_vec, y_train)

        # Predictions
        y_pred = model.predict(X_test_vec)

        # Evaluation
        print("Accuracy:", accuracy_score(y_test, y_pred))
        print("\nClassification Report:\n", classification_report(y_test, y_pred))
        print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

        # Example prediction
        sample_messages = [
            "Congratulations! You've won a $1000 Walmart gift card. Click here to claim.",
            "Hey, are we still meeting for lunch today?"
        ]
        sample_vec = vectorizer.transform(sample_messages)
        predictions = model.predict(sample_vec)
        for msg, pred in zip(sample_messages, predictions):
            print(f"Message: {msg}\nPrediction: {'SPAM' if pred == 1 else 'HAM'}\n")

    except Exception as e:
        print("Error:", e)
```

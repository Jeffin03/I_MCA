```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore")

def load_and_prepare_data():
    """
    Loads the MNIST digits dataset from sklearn and prepares it for training.
    Returns: X_train, X_test, y_train, y_test
    """
    try:
        digits = datasets.load_digits()  # 8x8 images of digits 0-9
        X = digits.data
        y = digits.target

        # Normalize features for better SVM performance
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        # Split into training and testing sets
        return train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    except Exception as e:
        print(f"Error loading dataset: {e}")
        exit(1)

def train_svm(X_train, y_train):
    """
    Trains an SVM classifier with RBF kernel.
    Returns: trained model
    """
    try:
        svm_clf = SVC(kernel='rbf', gamma=0.05, C=5)  # Tuned parameters
        svm_clf.fit(X_train, y_train)
        return svm_clf
    except Exception as e:
        print(f"Error training SVM: {e}")
        exit(1)

def evaluate_model(model, X_test, y_test):
    """
    Evaluates the trained SVM model and prints metrics.
    """
    try:
        y_pred = model.predict(X_test)
        print("\nModel Evaluation:")
        print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred))
        print("Confusion Matrix:")
        print(confusion_matrix(y_test, y_pred))
    except Exception as e:
        print(f"Error during evaluation: {e}")

if __name__ == "__main__":
    # Load and prepare data
    X_train, X_test, y_train, y_test = load_and_prepare_data()

    # Train SVM model
    model = train_svm(X_train, y_train)

    # Evaluate model
    evaluate_model(model, X_test, y_test)

```
